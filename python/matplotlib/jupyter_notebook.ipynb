# Matplotlib test

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
from scipy.stats import norm
import itertools
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

x = np.linspace(-6,6,1000)

ax.plot(x,norm.pdf(x, loc=0.0, scale=1.0), color='black',  linestyle='solid')
ax.plot(x,norm.pdf(x, loc=0.0, scale=0.5), color='black',  linestyle='dashed')
ax.plot(x,norm.pdf(x, loc=0.0, scale=0.25), color='black', linestyle='dashdot')

ax.set_title('First line plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
#fig.show()
```

```python
plt.rcParams['figure.figsize']= 5,5
X= np.linspace(0,5,20)
plt.scatter(X, 2.0*X+np.random.uniform(size=len(X)), color='blue', label='random')
X= np.linspace(0,5,1000)
plt.plot(X, 2.0*X, color='blue', linewidth=3, label='linear')

plt.title('Test')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(left=-3)
plt.ylim(bottom=-3)
plt.legend()
set_matplotlib_formats('png')
plt.show()
```

```python
plt.rcParams['figure.figsize']= 5,5
groups= ['a','b']
x_data= ['x1','x2','x3']
y_data= {'a': [1,2,3],'b':[3,2,1]}
y_err= {'a': [1,1,1],'b':[2,1,0.2]}

bar_loc= np.arange(len(x_data))
bar_w= 0.15
bar_offset= [-bar_w/2,bar_w/2]
plt.bar(bar_loc+bar_offset[0], y_data['a'], width=bar_w, yerr=y_err['a'], label='a')
plt.bar(bar_loc+bar_offset[1], y_data['b'], width=bar_w, yerr=y_err['b'], alpha=0.5, label='b')

plt.xlabel('x')
plt.ylabel('y')
plt.xticks(bar_loc, x_data)
plt.legend(loc='upper left')
plt.show()
```

# Scatter

```python
X0= np.linspace(0,5,1000)
Y0= 2.0*X0
X1= np.linspace(0,5,20)
Y1= 2.0*X1+np.random.uniform(-1.,1.,size=len(X1))

fig= plt.figure(figsize=(5,5))
ax= fig.add_subplot(1,1,1,title='Scatter',xlabel='x',ylabel='y')
ax.plot(X0, Y0, color='red', linewidth=0.5, label='true')
ax.scatter(X1, Y1, color='blue', label='noise')
ax.legend()
set_matplotlib_formats('png')
```

```python
X0= np.linspace(0,5,1000)
Y0= 2.0*X0
X1= np.linspace(0,5,20)
Y1= 2.0*X1+np.random.uniform(-1.,1.,size=len(X1))
X2= np.linspace(0,5,10)
Y2= 2.0*X2+np.random.uniform(-3.,3.,size=len(X2))

markers= ['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', '8', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_']
m1,m2= np.random.choice(range(len(markers)),2)
fig= plt.figure(figsize=(5,5))
ax= fig.add_subplot(1,1,1,title='Scatter',xlabel='x',ylabel='y')
ax.plot(X0, Y0, color='red', linewidth=0.5, label='true')
ax.scatter(X1, Y1, color='blue', label=f'noise1(marker={markers[m1]})', marker=markers[m1], s=64)
ax.scatter(X2, Y2, color='green', label=f'noise2(marker={markers[m2]})', marker=markers[m2], s=64)
ax.legend()
set_matplotlib_formats('png')
```

```python
f= lambda X: 2.0*X
X0= np.linspace(0,5,1000)
X1= np.linspace(0,5,20)
Y1= np.random.uniform(-1.,1.,size=len(X1))
X2= np.linspace(0,5,10)
Y2= np.random.uniform(-3.,3.,size=len(X2))

fig= plt.figure(figsize=(5,5))
ax= fig.add_subplot(1,1,1,title='Scatter',xlabel='x',ylabel='y')
ax.plot(X0, f(X0), color='red', linewidth=0.5, label='true')
ax.scatter(X1, f(X1)+Y1, color='blue', label=f'noise1', marker='x', s=64)
ax.scatter(X2, f(X2)+Y2, color='green', label=f'noise2', marker='*', s=64)

for i in range(len(X1)): ax.annotate(f'{Y1[i]:.1f}',(0.15+X1[i],f(X1[i])+Y1[i]),  #, rotation=-90
                                     verticalalignment='center', fontsize=10, color='blue')

for i in range(len(X2)): ax.annotate(f'{Y2[i]:.1f}',(X2[i],0.15+f(X2[i])+Y2[i]),
                                     horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='green')

ax.legend()
plt.show()
```

# Histogram

```python
X= np.concatenate((np.random.uniform(0,10,size=200), np.random.uniform(4,7,size=100)))

fig= plt.figure(figsize=(10,5))
ax1= fig.add_subplot(1,2,1,title='Histogram test',xlabel='x',ylabel='probability')
n,bins,patches= ax1.hist(X, bins=50, density=True, facecolor='g', alpha=0.75, label='hist')
ax1.legend()

ax2= fig.add_subplot(1,2,2,title='Histogram test',xlabel='x',ylabel='frequency')
n,bins,patches= ax2.hist(X, bins=25, facecolor='g', alpha=0.75, label='hist')
ax2.legend()

plt.show()
```

```python
X= np.concatenate((np.random.uniform(0,10,size=200), np.random.uniform(4,7,size=100)))
Y= np.concatenate((np.random.uniform(0,10,size=100), np.random.uniform(0,4,size=100), np.random.uniform(7,10,size=100)))

fig= plt.figure(figsize=(10,5))
ax1= fig.add_subplot(1,2,1,title='Histogram test',xlabel='x',ylabel='frequency')
ax1.hist(X, bins=50, color='b', alpha=0.5, label='X')
ax1.hist(Y, bins=50, color='r', alpha=0.5, label='Y')
ax1.legend()

ax2= fig.add_subplot(1,2,2,title='Histogram test',xlabel='x',ylabel='frequency')
ax2.hist([X,Y], bins=25, color=['b','r'], alpha=0.5, label=['X','Y'])
ax2.legend()

plt.show()
```
